```python
im = Image.open('../imgs/input.tif')
imarray = np.array(im)
```

```python
divisors = find_divisors(imarray.size)

print(imarray.shape)
print(imarray.size)
print("Possible values of x (divisors of imarray.size):")
print(divisors)
```

```python
# Compress message in imarray using Huffman code 

mx = 25
Ln = []  # largo promedio
Tc = []  # tasa de compresion
for x in range(1,mx):
    cimarray, tr = build_huffman_tree(imarray, x)
    Ln.append(len(cimarray) / imarray.size)
    Tc.append(1/Ln[-1])

print(Ln)
print(Tc)

plt.figure(figsize=(15, 5))

# Plot Ln
plt.subplot(1, 2, 1)
plt.plot(range(1, mx), Ln, 'b-o')
plt.xlabel('x')
plt.ylabel('Ln')
plt.title('Longitud promedio del simbolo vs x')
plt.grid(True)
plt.xticks(range(1, mx))  # Show only integer values on x-axis

# Plot Tc
plt.subplot(1, 2, 2)
plt.plot(range(1, mx), Tc, 'r-o')
plt.xlabel('x')
plt.ylabel('Tc')
plt.title('Tasa de compresion vs x')
plt.grid(True)
plt.xticks(range(1, mx))  # Show only integer values on x-axis

plt.tight_layout()
plt.show()
```
